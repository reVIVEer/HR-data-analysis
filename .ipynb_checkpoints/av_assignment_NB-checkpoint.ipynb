<font size="3"><b>Categorical Features:</b></font><br>
A categorical variable is one that has two or more categories and each value in that feature can be categorised by them. For example, gender is a categorical variable having two categories (male and female). Now we cannot sort or give any ordering to such variables. They are also known as Nominal Variables.

Categorical Features in the dataset: **department, region, education, gender, recruitment_channel, KPIs_met >80%, awards_won**

<font size="3"><b>Continuous Feature:</b></font><br />
A feature is said to be continous if it can take values between any two points or between the minimum or maximum values in the features column.

Continous Features in the dataset: **no_of_trainings, age, previous_year_rating, length_of_service, avg_training_score**

<font size="4">Training Score vs Number of Trainings</font>

```python
#Group the entries by department, then get how many entries are there; i.e. the number of trainings
temp1 = df2.groupby(['department']).count()
temp1 = temp1.reset_index()

#Do the same, but sum the values to get the total values of everything by year.
temp2 = df2.groupby(['department']).sum()
temp2 = temp2.reset_index()

#Normalize the data, i.e. zero mean and unit std. I did this to be able to compare the shapes of both graphs, since 
#they have different ranges
normalised_df = pd.DataFrame()

normalised_df['no_of_trainings'] = temp1['no_of_trainings']
normalised_df['Average Training Score'] = temp2['avg_training_score']
normalised_df = (normalised_df - normalised_df.mean()) / normalised_df.std()#(normalised_df.max() - normalised_df.min()) 
normalised_df['Department'] = temp1['department']


#Plot
plt.figure(figsize=(15, 9))
ax = sns.pointplot(x = normalised_df.Department, y = normalised_df.no_of_trainings, color = 'blue', label='Number of Training')
ax = sns.pointplot(x = normalised_df.Department, y = normalised_df['Average Training Score'], color = 'red', label='Average Training Score')

blue_patch = mpatches.Patch(color='blue', label='Number of Training')
red_patch = mpatches.Patch(color='red', label='Average Training Score')
plt.legend(handles=[blue_patch, red_patch], loc='upper left', fontsize = 16)

plt.xticks(rotation=45);
```

The shapes follow each other well, there wasn't any weird high score with little number of trainings or vice versa.

<font size="4">KPI Met vs Awards Won</font>

```python
#Group the entries by department, then get how many entries are there; i.e. the number of trainings
temp1 = df2.groupby(['department']).count()
temp1 = temp1.reset_index()

#Do the same, but sum the values to get the total values of everything by year.
temp2 = df2.groupby(['department']).sum()
temp2 = temp2.reset_index()

#Normalize the data, i.e. zero mean and unit std. I did this to be able to compare the shapes of both graphs, since 
#they have different ranges
normalised_df = pd.DataFrame()

normalised_df['KPIs_met >80%'] = temp1['KPIs_met >80%']
normalised_df['Awards Won'] = temp2['awards_won?']
normalised_df = (normalised_df - normalised_df.mean()) / normalised_df.std()#(normalised_df.max() - normalised_df.min()) 
normalised_df['Department'] = temp1['department']


#Plot
plt.figure(figsize=(15, 9))
ax = sns.pointplot(x = normalised_df.Department, y = normalised_df['KPIs_met >80%'], color = 'blue', label='Number of Training')
ax = sns.pointplot(x = normalised_df.Department, y = normalised_df['Awards Won'], color = 'red', label='Awards Won')

blue_patch = mpatches.Patch(color='blue', label='KPI Met > 80%')
red_patch = mpatches.Patch(color='red', label='Awards Won')
plt.legend(handles=[blue_patch, red_patch], loc='upper left', fontsize = 16)

plt.xticks(rotation=45);
```

<font size="4">Department Training Score</font>

I want to see how different departments perform according to different metrics. The ones I chose are:
* Number of Trainings
* Training Score
* Average Training Score per Employee

```python
#First is the number of trainings per department, second is the training score per department, third is the average training score per employee per department
department_difference_metric = [df2.department.value_counts().index, df2.groupby(['department']).sum()['avg_training_score'].sort_values(ascending=False).index, df2.groupby(['department']).mean()['avg_training_score'].sort_values(ascending=False).index]

#Dataframe to be used for plotting.
department_evolution_df = pd.DataFrame(columns=['department', 'rank_type', 'rank'])

#Populate the dataframe
for metric in range(3):
    for department in range(len(department_difference_metric[metric])):
        department_evolution_df = department_evolution_df.append({'department':department_difference_metric[metric][department], 'rank_type': metric, 'rank':department},
                                   ignore_index=True)


fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111)

sns.pointplot(x=department_evolution_df.rank_type,
              y=9-department_evolution_df['rank'], 
              hue=department_evolution_df.department)

for i in range(len(department_difference_metric[0])):
    ax.text(-0.75, 9-i, department_difference_metric[0][i], fontsize=11)
    ax.text(2.1, 9-i, department_difference_metric[2][i], fontsize=11)
    
ax.set_xlim([-2,4])

xs = [0.0, 1.0, 2.0]
x_labels = ['Number of Trainings', 'Training Score', 'Avg Training Score']
plt.xticks(xs, x_labels, rotation='vertical')

ax.set_xlabel('Training Score Metric')

ys = range(1,10)
y_labels = ['9th', '8th', '7th', '6th', '5th', '4th', '3rd', '2nd', '1st']
plt.yticks(ys, y_labels)
ax.set_ylabel('Department Rank')

plt.legend(fontsize = 15)

plt.show();
```

## ML Model

```python
#Machine Learning Algorithm (MLA) Selection and Initialization
# Uncomment the ML algorithms to compare all of them
MLA = [
    #Ensemble Methods
    ensemble.AdaBoostClassifier()
#     ensemble.BaggingClassifier(),
#     ensemble.ExtraTreesClassifier(),
#     ensemble.GradientBoostingClassifier(),
#     ensemble.RandomForestClassifier(),

#     #Gaussian Processes
#     gaussian_process.GaussianProcessClassifier(),
    
#     #GLM
#     linear_model.LogisticRegressionCV(),
#     linear_model.PassiveAggressiveClassifier(),
#     linear_model.RidgeClassifierCV(),
#     linear_model.SGDClassifier(),
#     linear_model.Perceptron(),
    
#     #Navies Bayes
#     naive_bayes.BernoulliNB(),
#     naive_bayes.GaussianNB(),
    
#     #Nearest Neighbor
#     neighbors.KNeighborsClassifier(),
    
#     #SVM
#     svm.SVC(probability=True),
#     svm.LinearSVC(),
    
#     #Trees    
#     tree.DecisionTreeClassifier(),
#     tree.ExtraTreeClassifier()   
    ]



#split dataset in cross-validation with this splitter
cv_split = model_selection.ShuffleSplit(n_splits = 10, test_size = .3, train_size = .6, random_state = 0 ) # run model 10x with 60/30 split intentionally leaving out 10%

#create table to compare MLA metrics
MLA_columns = ['MLA Name', 'MLA Parameters','MLA Train Accuracy Mean', 'MLA Test Accuracy Mean', 'MLA Test Accuracy 3*STD' ,'MLA Time']
MLA_compare = pd.DataFrame(columns = MLA_columns)

#create table to compare MLA predictions
MLA_predict = df[Target]

#index through MLA and save performance to table
row_index = 0
for alg in MLA:

    #set name and parameters
    MLA_name = alg.__class__.__name__
    MLA_compare.loc[row_index, 'MLA Name'] = MLA_name
    MLA_compare.loc[row_index, 'MLA Parameters'] = str(alg.get_params())
    
    #score model with cross validation:
    cv_results = model_selection.cross_validate(alg, df[df1_x_bin], df[Target], cv  = cv_split, return_train_score=True)

    MLA_compare.loc[row_index, 'MLA Time'] = cv_results['fit_time'].mean()
    MLA_compare.loc[row_index, 'MLA Train Accuracy Mean'] = cv_results['train_score'].mean()
    MLA_compare.loc[row_index, 'MLA Test Accuracy Mean'] = cv_results['test_score'].mean()   
    #if this is a non-bias random sample, then +/-3 standard deviations (std) from the mean, should statistically capture 99.7% of the subsets
    MLA_compare.loc[row_index, 'MLA Test Accuracy 3*STD'] = cv_results['test_score'].std()*3   #let's know the worst that can happen!
    

    #save MLA predictions - see section 6 for usage
    alg.fit(df[df1_x_bin], df[Target])
    MLA_predict[MLA_name] = alg.predict(df[df1_x_bin])
    row_index+=1

    
#print and sort table:
MLA_compare.sort_values(by = ['MLA Test Accuracy Mean'], ascending = False, inplace = True)
MLA_compare
#MLA_predict
```
